```python
%matplotlib inline

import matplotlib.pylab as plt
import numpy as np
from sympy import *
import pandas as pd
```

```python
var("x,t")

plot (sin(x),(x,0,pi))
integrate(sin(x),(x,0,pi))
```

```python
N= 1000
f=lambda x: np.sin(x)
a,b = 0,np.pi
Altura = 1

x=np.random.uniform(a,b,N)
y=np.random.uniform(a,Altura,N)
X=np.arange (a,b,.01)
fx=f(x)

plt.plot(x[y>fx],y[y>fx], 'go')
plt.plot(x[y<fx],y[y<fx], 'ro')
plt.plot(X,f(X))

N_0=len(x[y<fx])
Base=b-a
yest=N_0/N*(Base*Altura)
print (yest)
```
